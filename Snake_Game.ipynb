```python
import pygame
import random
import time
import sys
import xlsxwriter
from datetime import datetime

# Initialize Pygame
pygame.init()

# Screen dimensions
width, height = 600, 600
snake_x = width // 2
snake_y = height // 2
change_x, change_y = 0, 0

# Create game window
game_screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("SNAKE GAME")

# Food position
food_x, food_y = random.randint(0, width // 10) * 10, random.randint(0, height // 10) * 10

# Snake body initialization
snake_body = [(snake_x, snake_y)]

# Load background image
background = pygame.image.load("C:/Users/rajuc/Snake and Apple Game/block.jpg").convert()
background = pygame.transform.scale(background, (width, height))

# Load sound for eating food
sound = pygame.mixer.Sound("C:/Users/rajuc/Snake and Apple Game/1_snake_game_resources_crash.mp3")

# Background music
pygame.mixer.music.load("C:/Users/rajuc/Snake and Apple Game/retro-game-music-245230.mp3")
music_started = False

# Score initialization
score = 0

# Variable to check if game is paused
is_paused = False

# Function to save the score to an Excel file
def save_score_to_excel(name, score):
    try:
        # Open the workbook in append mode
        workbook = xlsxwriter.Workbook("C:/Users/rajuc/OneDrive/Desktop/Excel Sheets/Snake_Game.xlsx")
        worksheet = workbook.add_worksheet("Sheet")

        # Check if the sheet already exists and find the last row used
        try:
            worksheet = workbook.get_worksheet_by_name('Sheet')
            last_row = len(worksheet.get_row(1)) + 1  # Find next empty row
        except:
            # Add the header row if the sheet doesn't exist
            worksheet.write(0, 0, "SI.NO")
            worksheet.write(0, 1, "Name")
            worksheet.write(0, 2, "Date")
            worksheet.write(0, 3, "Time")
            worksheet.write(0, 4, "Score")
            last_row = 1

        # Adding the new score to the next row
        current_date = datetime.now().strftime("%Y-%m-%d")
        current_time = datetime.now().strftime("%H:%M:%S")

        worksheet.write(last_row, 0, last_row)  # Incremented SI.NO
        worksheet.write(last_row, 1, name)
        worksheet.write(last_row, 2, current_date)
        worksheet.write(last_row, 3, current_time)
        worksheet.write(last_row, 4, score)

        workbook.close()
        print("Score saved successfully!")

    except Exception as e:
        print(f"Error while saving score: {e}")


# Game over function
def game_over():
    global score
    font = pygame.font.SysFont('TIMEROMAN', 50)
    game_over_text = font.render("GAME OVER", True, (255, 0, 0))
    game_screen.blit(game_over_text, (width // 2 - game_over_text.get_width() // 2, height // 2))
    pygame.display.update()
    time.sleep(1)
    save_score_to_excel("Narasimha Raju", score)  # Score display
    pygame.quit()
    sys.exit()

# Display Score
def display_score():
    font = pygame.font.SysFont('TIMEROMAN', 30)
    score_display = font.render(f"Score: {score}", True, (250, 0, 0))
    game_screen.blit(score_display, (10, 10))

# Snake movement
def display_snake():
    global snake_x, snake_y, food_x, food_y, score, snake_body

    snake_x = (snake_x + change_x) % width
    snake_y = (snake_y + change_y) % height

    # Check if the snake eats the food
    if food_x == snake_x and food_y == snake_y:
        food_x, food_y = random.randint(0, width)//10 * 10, random.randint(0, height)//10 * 10
        sound.play()
        score += 1  # score when food is eaten
    else:
        del snake_body[0]

    # Add new head to the snake body
    snake_body.append((snake_x, snake_y))

    # Check if snake collides with itself
    if (snake_x, snake_y) in snake_body[:-1]:
        sound.play()
        game_over()

    # Draw everything
    game_screen.blit(background, (0, 0))

    # Draw snake
    for x, y in snake_body:
        pygame.draw.rect(game_screen, (255, 255, 255), [x, y, 10, 10])

    # Draw food
    pygame.draw.rect(game_screen, (0, 255, 0), [food_x, food_y, 10, 10])

    # Display updated score
    display_score()
    pygame.display.update()

def display_pause():
    font = pygame.font.SysFont('TIMEROMAN', 50)
    pause_text = font.render("PAUSED", True, (255, 255, 0))
    game_screen.blit(pause_text, (width // 2 - pause_text.get_width() // 2, height // 2))
    pygame.display.update()
    music_started=False

# Gaming loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if not music_started:
                pygame.mixer.music.play(-1, 0.0)
                music_started = True

            if event.key == pygame.K_LEFT and change_x == 0:
                change_x = -10
                change_y = 0
            elif event.key == pygame.K_RIGHT and change_x == 0:
                change_x = 10
                change_y = 0
            elif event.key == pygame.K_UP and change_y == 0:
                change_y = -10
                change_x = 0
            elif event.key == pygame.K_DOWN and change_y == 0:
                change_y = 10
                change_x = 0

            # Pause and resume functionality
            if event.key == pygame.K_p:  # 'P' key to pause/unpause
                is_paused = not is_paused
                if is_paused:
                    pygame.mixer.music.pause()  
                else:
                    pygame.mixer.music.unpause()  

            

    if is_paused:
        display_pause()
    else:
        display_snake()
    
    pygame.time.Clock().tick(12)

pygame.quit()
```
